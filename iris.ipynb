# Importing header files

```python
import pandas as pd
from sklearn.model_selection import train_test_split
import numpy as np
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score, precision_score, recall_score, f1_score
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
```

# Loading data

```python
df=pd.read_csv('D:\OIB\\task1\iris.csv')
class_mapping = {'Iris-setosa': 0, 'Iris-versicolor': 1, 'Iris-virginica': 2}
df['Species'] = df['Species'].map(class_mapping)
print(df.shape)
print(df.head())

X = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

# Visualizing data

```python
data = df.iloc[1:,1:]
pairplot = sns.pairplot(data, hue="Species", diag_kind="kde")
pairplot.add_legend()
plt.show()
```

# Splitting and feature scaling

```python
scaler = StandardScaler()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=None)
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

# Training the KNN model and testing the classification

```python
knn = KNeighborsClassifier(n_neighbors=5)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print(accuracy*100)
```
